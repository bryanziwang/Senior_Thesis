# AniWave

```python
from selenium import webdriver
import undetected_chromedriver as uc 
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
import time 
import numpy as np
import pandas as pd

start_time = time.time()
 
# Initialize driver
options = uc.ChromeOptions() 
options.add_argument("--disable-extensions")
options.add_argument('--disable-notifications')
options.add_argument('--incognito')

options.headless = True  # Set headless to False to run in non-headless mode
driver = uc.Chrome(use_subprocess=True, options=options) 


# ------------ LINKS ------------ #
# Get the link for each page of shows (1-390)
list_webpages = []
for page_number in range(300,391):
    list_webpages = np.append(list_webpages, f'https://aniwave.to/az-list?page={page_number}')

# For each page of shows, get the link for each show displayed
links = []  
for list_webpage in list_webpages:
    
    try:
        driver.set_page_load_timeout(20)
        driver.get(list_webpage)
    except Exception as e:
        continue
    
    # LINKS: Find the parent elements by class name
    try:
        driver.set_page_load_timeout(20)
        parent_elements = driver.find_elements('class name', 'ani')
    except Exception as e:
        continue
    
    # LINKS: Find the show link and append it to the array of links
    for parent_element in parent_elements[1:]: # Seems to duplicate the first show on each page
        try: 
            driver.set_page_load_timeout(20)
            link_element = parent_element.find_element('tag name', 'a')
            link_url = link_element.get_attribute('href')
            links = np.append(links, link_url)
        except Exception as e:
            continue    

    
# ------------ SCRAPE DETAILS ------------ #

# Scrape the following details about each show
names = []
types = []
ratings = []
vote_counts = []
start_dates = []
studios = []

batch_size = 1000

# Iterate through each link to scrape the rating
for link_url in links:
    # Visit the link
    try:
        driver.set_page_load_timeout(20)
        driver.get(link_url)
    except Exception as e:
        continue
    
    # Scrape the name
    try: 
        driver.set_page_load_timeout(20)
        info = driver.find_element('id', 'w-info')
        parent_element = info.find_element('class name', 'd-title')
        name = parent_element.text
        names = np.append(names, name)
    except Exception as e:
        names = np.append(names, None)
    
    
    # Scrape the show type
    try: 
        driver.set_page_load_timeout(20)
        parent_element = driver.find_element('xpath', "//div[@class='bmeta']//div[@class='meta'][1]//div[1]//span//a")
        type_ = parent_element.text
        types = np.append(types, type_)
    except Exception as e:
        types = np.append(types, None)
    
    # Scrape the rating
    try: 
        driver.set_page_load_timeout(20)
        parent_element = driver.find_element('css selector', "span[itemprop='ratingValue']")
        rating = parent_element.text
        ratings = np.append(ratings, rating)
    except Exception as e:
        ratings = np.append(ratings, None)
    
    # Scrape the count of votes
    try: 
        driver.set_page_load_timeout(20)
        parent_element = driver.find_element('css selector', "span[itemprop='reviewCount']")
        vote_count = parent_element.text
        vote_counts = np.append(vote_counts, vote_count)
    except Exception as e:
        vote_counts = np.append(vote_counts, None)
    
    # Scrape the start date
    try: 
        driver.set_page_load_timeout(20)
        parent_element = driver.find_element('css selector', "span[itemprop='dateCreated']")
        start_date = parent_element.text
        start_dates = np.append(start_dates, start_date)
    except Exception as e:
        start_dates = np.append(start_dates, None)
    
    # Scrape the studio
    try: 
        driver.set_page_load_timeout(20)
        parent_element = driver.find_element('css selector', "a[itemtype='https://schema.org/Person']")
        child_element = parent_element.find_element('css selector', "span[itemprop='name']")
        studio = child_element.text
        studios = np.append(studios, studio)
    except Exception as e:
        studios = np.append(studios, None)
        print(f"Error: {e}")
        print(name)
        
    # Check if the batch size is reached
    if len(names) >= batch_size:
        
        # combine into pandas dataframe
        scraped_data = pd.DataFrame({'Title':names, 'Type': types,'Rating':ratings, 'Vote Count':vote_counts, 
                                     'Start Date':start_dates, 'Studio':studios})
        
        # Write the data to a file
        with open('aniwave.csv', 'a') as file:
            scraped_data.to_csv(file, header=not file.tell(), index=False)
        
        # Clear the large dataframe for the next batch
        scraped_data = pd.DataFrame()
        
        names = []
        types = []
        ratings = []
        vote_counts = []
        start_dates = []
        studios = []
        
        
# Check if there are any remaining data
if len(names) != 0:
    
    # combine into pandas dataframe
    scraped_data = pd.DataFrame({'Title':names, 'Type': types,'Rating':ratings, 'Vote Count':vote_counts, 
                                 'Start Date':start_dates, 'Studio':studios})
    
    with open('aniwave.csv', 'a') as file:
        scraped_data.to_csv(file, header=not file.tell(), index=False)


driver.quit()

end_time = time.time()
elapsed_time = end_time - start_time
print(f'Time elapsed: {elapsed_time} seconds')
```
